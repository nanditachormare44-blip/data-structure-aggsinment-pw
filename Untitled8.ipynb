```python
#write a program to demostrate encapsulation by creating a class Bankaccount with private attributes balance and methods to deposit withdraw and check balance.

class Bankaccount:
  def __init__(self):
    self.__balance = 0

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
    else:
      print("Amount must be positive.")

  def withdrawl(self, amount):
    if amount > 0 and self.__balance >= amount:
      self.__balance -= amount
    else:
      print("Invalid amount or insufficient balance.")

  def check_balance(self):
    return self.__balance
```

```python
#demonstrate operator overloading by creating a class vector and ovveriding the add method to add two vectors.
class vector:
  def __init__(self,g,h):
    self.g =g
    self.h =h

  def __add__(self,other):
    return vector(self.g + other.g, self.h + other.h)

  def __str__(self):
    return f"({self.g},{self.h})"

v1 = vector(1,2)
v2 = vector(3,4)
result = v1 + v2
print(result)
```
